# 5. Summary and Key Findings

This project analyzed the FIFA 2017 player dataset to understand how player performance relates to age, physical attributes, and position. The dataset was cleaned by handling missing values, standardizing variables, and creating new features to improve the quality and consistency of the analysis.

Descriptive statistics showed that player ratings are concentrated in the mid-range. Grouped analysis and correlation analysis revealed that player rating is positively related to attributes such as speed and strength, but no single variable fully explains overall performance.

Feature exploration and visualizations highlighted key patterns across the dataset. Player ratings tend to increase with experience, reach a peak during prime years, and slightly decline at older ages. Speed decreases with age, while strength remains more stable over time. Additionally, positional analysis showed that wide attacking and midfield roles tend to have the highest average speed, reflecting the physical demands of those positions.


## What We Set Out to Answer

Using a dataset of over 17,500 FIFA-registered professional football players, our team explored the following questions:

1. How do player physical attributes (speed and strength) change with age? (Line Plot)
2. Which club positions have the highest average speed? (Bar Plot)
3. What is the distribution of player ratings? (Histogram)
4. What is the relationship between dribbling ability and ball control? (Scatter Plot)

## Key Findings

### 1. Player ratings change across a player’s career
Player ratings tend to increase as players gain experience, peak during their prime years, and slightly decline at older ages. This reflects the balance between skill development and physical changes over time.

### 2. Most players fall within the mid-range of ratings
The distribution of ratings is centered around the middle range (approximately 60–75), with relatively few players reaching elite levels. This suggests that high-performing players are less common in the dataset.

### 3. Speed declines with age while strength remains stable
Speed shows a negative relationship with age, indicating that players tend to lose speed as they get older. In contrast, strength remains more consistent across age groups, suggesting different physical attributes change at different rates.

### 4. Player rating is influenced by multiple attributes
Correlation analysis shows that rating is positively related to both speed and strength. However, these relationships are moderate, indicating that overall performance depends on a combination of factors rather than a single dominant attribute.

### 5. Position plays a key role in physical attributes
Wide attacking and midfield positions have the highest average speed, reflecting the need for movement and agility in those roles. This highlights how positional responsibilities influence player characteristics.


## Limitations and Next Steps

- **Data source:** FIFA ratings are generated by game developers and may not perfectly reflect real-world performance.
- **Missing national positions:** Due to the fact that not every player competes in the national league, over 94% of players had no national team position, limiting international comparisons.
- **No temporal data:** The dataset is a single snapshot, we cannot track how individual players improve or decline over time.

A natural next step would be to compare multiple years of FIFA data to study career trajectories, or to merge with real match performance statistics (e.g., goals, assists, pass accuracy) to validate whether FIFA ratings align with actual output on the pitch.
